```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error

pd.set_option('display.max_columns', None)
```

```python
df0 = pd.read_csv("train.csv")
df1 = pd.read_csv("test.csv")
```

```python
df0["is_train"] = 1
df1["is_train"] = 0
```

```python
df = pd.concat([df0 , df1], ignore_index=True)
```

```python
df.head()
```

```python
columns_to_be_deleted = ['Street','Alley','LandContour','Utilities','LandSlope','Condition2','MasVnrType','HeatingQC','2ndFlrSF','LowQualFinSF','BsmtHalfBath','KitchenAbvGr','GarageQual','MiscVal','MiscFeature','PoolArea','3SsnPorch']
```

```python
len(columns_to_be_deleted)
```

```python
df.drop(columns = columns_to_be_deleted, inplace = True, axis=1)
```

```python
df.head()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print(len(categorical_cols), "categorical columns")
print(categorical_cols)
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
df.head()
```

```python
for col in df.select_dtypes(include='bool').columns:
    df[col] = df[col].astype(int)
```

```python
# Calculate the ratio of True values for boolean columns
bool_cols = df.select_dtypes(include='int').columns[df.select_dtypes(include='int').apply(lambda x: x.isin([0, 1]).all())]
bool_ratios = df[bool_cols].mean()
```

```python
columns_to_drop = []
for col, ratio in bool_ratios.items():
  if ratio > 0.95 or ratio < 0.05:
    columns_to_drop.append(col)

# Drop the identified columns from the DataFrame
df.drop(columns=columns_to_drop, inplace=True, axis=1)
```

```python
df.head()
```

```python
len(df.columns)
```

```python

nan_cols = df.columns[df.isna().any()]
print(nan_cols)
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
```

```python
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
```

```python
bsmt_cols = ['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF']
df[bsmt_cols] = df[bsmt_cols].fillna(0)
```

```python
df['BsmtFullBath'] = df['BsmtFullBath'].fillna(0)
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
garage_cols = ['GarageCars','GarageArea']
df[garage_cols] = df[garage_cols].fillna(0)
```

```python
col = df.pop('SalePrice')
df['SalePrice'] = col
```

```python
col = df.pop('Id')
df['Id'] = col
```

```python
train_df = df[df['is_train'] == 1]
test_df = df[df['is_train'] == 0]
```

```python
train_df.head()
```

```python
train_df = train_df.drop(columns=["is_train"])
test_df = test_df.drop(columns=["is_train"])
```

```python
X_train = train_df[train_df.columns[:-2]].values
y_train = train_df['SalePrice'].values
```

```python
print(y_train)
```

```python
x_test = test_df[test_df.columns[:-2]].values
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
x_test = scaler.transform(x_test)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

```python
test_df.head()
```

```python
print(len(test_df))
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

```python
submission = pd.DataFrame({'Id': test_df['Id'], 'SalePrice': y_pred})
submission.to_csv('submission.csv', index=False)
```
